```python
def clean_columns(df: pd.DataFrame) -> pd.DataFrame:
    """
  This  function receives a DF, cleans
  and modifies some column names.

  Return
  DF
    """
    df_modify = df.copy()
# mayusculas
    df_modify['keyword'] = df_modify['keyword'].str.title()
#modificar la columna location
    df_modify['location'] = df_modify['location'].str.lower()
    df_modify['location'] = df_modify['location'].str.title()
    #ahora creamos una nueva columna para ver las diferencias
    df_modify['location_clean'] = df_modify['location'].replace(to_replace=["Los Angeles", " LA", "Los Angeles, Califnordia", "Los Angeles, Ca"],
                                                           value="Los Angeles, California") 
    #remplazamos los caracteres especiales, pero mantuvo con comas
    df_modify['location_clean'] = df_modify.select_dtypes('object')['location_clean'].apply(lambda x: re.sub('[^.,a-zA-Z0-9 \n\.]', '', (x)))
    return df_modify
```
